```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python
df = pd.read_csv("2020 us dollar.csv")
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df = df.set_index("Date")
df.head()
```

```python
mpf.plot(df, type = "candle", tight_layout = True, figratio = (20, 12), title = "US dollar Index 2020", style = "yahoo")
```

```python
mpf.plot(df["2020-03"], type = "candle", volume = True)
```
