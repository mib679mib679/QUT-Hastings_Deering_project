```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("HD_processed.csv")
df.info()
print(df.shape)
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# number of transactions from monthly view
df = df[df.Year ==2020]
print(df.shape)
df['Month'] = df['Month'].astype("int64")
df["Month"] = df["Month"].sort_values(ascending=True, ignore_index=True)
df["Month"].value_counts(sort=False).plot(kind="bar",figsize=(12, 6))
```

```python
df = df[df.Year ==2021]
print(df.shape)
df['Month'] = df['Month'].astype("int64")
df["Month"] = df["Month"].sort_values(ascending=True, ignore_index=True)
df["Month"].value_counts(sort=False).plot(kind="bar",figsize=(12, 6))
```

```python
building = pd.read_csv("building_approvals.csv")
building.info()

building2019 = building[building.Year == 2019]
building2019 = building2019[["Month", "Number"]]
building2019 = building2019.sort_values("Month")

building2020 = building[building.Year == 2020]
building2020 = building2020[["Month", "Number"]]
building2020 = building2020.sort_values("Month")

building2021 = building[building.Year == 2021]
building2021 = building2021[["Month", "Number"]]
building2021 = building2021.sort_values("Month")

plt.subplots(figsize=(12,8))
y_position = np.arange(len(building2019["Month"]))
plt.xticks(y_position, building2019["Month"])
plt.title("Monthly number of building approvals in Toowoomba")
plt.ylabel("Number of building approvals")
plt.xlabel("Month")
plt.plot(y_position, building2019["Number"], label = 2019)
plt.plot(y_position, building2020["Number"], label = 2020)
plt.plot(y_position, building2021["Number"], label = 2021)
plt.legend()
```

```python
value = pd.read_csv("value_building_approvals.csv")
value.info()

value2019 = value[value.Year == 2019]
value2019 = value2019[["Month", "Value"]]
value2019 = value2019.sort_values("Month")

value2020 = value[value.Year == 2020]
value2020 = value2020[["Month", "Value"]]
value2020 = value2020.sort_values("Month")

value2021 = value[value.Year == 2021]
value2021 = value2021[["Month", "Value"]]
value2021 = value2021.sort_values("Month")

plt.subplots(figsize=(12,8))
y_position = np.arange(len(value2019["Month"]))
plt.xticks(y_position, value2019["Month"])
plt.title("Monthly value of building jobs in Toowoomba")
plt.ylabel("value of building jobs (AUD)")
plt.xlabel("Month")
#plt.plot(y_position, value2019["Value"], label = 2019)
#plt.plot(y_position, value2020["Value"], label = 2020)
#plt.plot(y_position, value2021["Value"], label = 2021)
plt.bar(y_position-0.2, value2019["Value"], width = 0.2, label = 2019)
plt.bar(y_position+0.005, value2020["Value"], width = 0.2, label = 2020)
plt.bar(y_position+0.2, value2021["Value"], width = 0.2, label = 2021)
plt.legend()
```

```python
PPI = pd.read_csv("PPI.csv")
PPI.info()

PPI2019 = PPI[PPI.Year == 2019]
PPI2019 = PPI2019[["Quarter", "Index"]]
PPI2019 = PPI2019.sort_values("Quarter")

PPI2020 = PPI[PPI.Year == 2020]
PPI2020 = PPI2020[["Quarter", "Index"]]
PPI2020 = PPI2020.sort_values("Quarter")

PPI2021 = PPI[PPI.Year == 2021]
PPI2021 = PPI2021[["Quarter", "Index"]]
PPI2021 = PPI2021.sort_values("Quarter")

plt.subplots(figsize=(12,8))
y_position = np.arange(len(PPI2019["Quarter"]))
plt.xticks(y_position, PPI2019["Quarter"])
plt.title("Input prices to house construction recorded strong increases in quarterly and annual terms")
plt.ylabel("Producer Price Index")
plt.xlabel("Quarter")
plt.plot(y_position, PPI2019["Index"], label = 2019)
plt.plot(y_position, PPI2020["Index"], label = 2020)
plt.plot(y_position, PPI2021["Index"], label = 2021)
plt.legend()
```

```python
df['Month'] = df['Month'].astype("object")
covid = df[df.Year == 2020]
covid = covid[["Month", "Monthly_Covid_cases_Toowoomba"]]
covid = covid.drop_duplicates(keep="first")
covid = covid.sort_values("Month")

covid2021 = df[df.Year == 2021]
covid2021 = covid2021[["Month", "Monthly_Covid_cases_Toowoomba"]]
covid2021 = covid2021.drop_duplicates(keep="first")
covid2021 = covid2021.sort_values("Month")

plt.subplots(figsize=(12,8))
y_position = np.arange(len(covid2021["Month"]))
plt.xticks(y_position, covid2021["Month"])
plt.title("Monthly Covid cases in Toowoomba")
plt.ylabel("Amount of Covid cases")
plt.xlabel("Month")
plt.bar(y_position, covid["Monthly_Covid_cases_Toowoomba"], label = 2020)
plt.bar(y_position, covid2021["Monthly_Covid_cases_Toowoomba"], label = 2021)
plt.legend()
```
