```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("HD_processed.csv")
df.info()
print(df.shape)
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df = df.drop('Unique_ID', 1)
df = df.drop('Invoice_Date', 1)
df = df.drop('Customer_Id', 1)
df = df.drop('Warehouse', 1)
df = df.drop('Postal_Code', 1)
df = df.drop('Sales_Order_Id', 1)
df = df.drop('Created_Date', 1)
df = df.drop('Job_Description', 1)
df = df.drop('Major_Minor_Class', 1)
df = df.drop('Real_Invoice_Date', 1)
df = df.drop('Quarter', 1)
df = df.drop('BigCustomer', 1)
df['Month'] = df['Month'].astype("object")
```

```python
df.info()
```

```python
df_dum = pd.get_dummies(df, columns = ['Customer_MarketSegment_Id', 'Source_Of_Supplier_Id', 'Sales_Channel', 'Month', 'Quantile_transactions'])
```

```python
df_dum.info()
```

```python
df_dum.corr(method='spearman')
```

```python
corr = df_dum.corr(method='spearman')  
fig = plt.figure(figsize=(14,12))  #
ax = fig.add_subplot(111)  #
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)  #
fig.colorbar(cax)   #BAR
ticks = np.arange(0,len(df_dum.columns),1)  #
ax.set_xticks(ticks)  #
plt.xticks(rotation=90)  
ax.set_yticks(ticks)  #
ax.set_xticklabels(df_dum.columns)  #X
ax.set_yticklabels(df_dum.columns)  #Y
plt.show()
```
